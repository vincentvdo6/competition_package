# Bagging Probe: Train on 85% Data Subsets
**Goal**: Each seed trains on a different random 85% of training sequences.
Data diversity should decorrelate predictions more than seed-only variation
(which gives mean correlation 0.947).

**Plan**: 12 screening seeds. Compare prediction correlation vs full-data base.
If mean pairwise correlation drops below 0.93 -> scale up and blend with base.

**Key**: Quality drop must be modest (val within -0.005 of base mean 0.2689).
10721 total train sequences x 0.85 = 9113 sequences per seed.

```python
# Cell 2: Train 12 bagged seeds (85% data subsets)
import os, subprocess, sys
os.chdir("/content/competition_package")

CONFIG = "configs/gru_parity_v1_bag85.yaml"
SEEDS = [80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90, 91]

print(f"=== BAGGING PROBE (85% data subsets) ===")
print(f"Config: {CONFIG}")
print(f"Seeds: {SEEDS}")
print(f"Each seed trains on ~9113/10721 sequences")
print("=" * 60, flush=True)

for seed in SEEDS:
    print(f"\n{'='*60}")
    print(f"Training bagged seed {seed}")
    print(f"{'='*60}", flush=True)
    proc = subprocess.Popen(
        [sys.executable, "-u", "scripts/train.py",
         "--config", CONFIG,
         "--seed", str(seed), "--device", "cuda"],
        stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True
    )
    for line in proc.stdout:
        print(line, end="", flush=True)
    proc.wait()
    if proc.returncode != 0:
        print(f"ERROR: seed {seed} failed with rc={proc.returncode}")

print(f"\nAll {len(SEEDS)} bagged seeds done!")
```

```python
# Cell 3: Evaluate bagged models
import os, glob, torch
os.chdir("/content/competition_package")

GRU_BASE_MEAN = 0.2689  # mean of 23 GRU base seeds (full data)

results = []
for pt in sorted(glob.glob("logs/gru_parity_v1_bag85_seed*.pt")):
    basename = os.path.basename(pt)
    if '_epoch' in basename:
        continue
    ckpt = torch.load(pt, map_location="cpu", weights_only=False)
    score = float(ckpt.get("best_score", 0))
    epoch = ckpt.get("best_epoch", "N/A")
    seed = int(basename.split("seed")[1].split(".")[0])
    results.append((seed, basename, score, epoch))

results.sort(key=lambda x: -x[2])

print(f"{'Rank':<6} {'Seed':<6} {'Val Score':>10} {'Epoch':>6}")
print("-" * 35)
for i, (seed, name, score, epoch) in enumerate(results, 1):
    print(f"{i:<6} s{seed:<5} {score:>10.4f} {str(epoch):>6}")

if results:
    scores = [s for _, _, s, _ in results]
    mean_val = sum(scores) / len(scores)
    delta = mean_val - GRU_BASE_MEAN
    print(f"\nBagged (85%): {len(results)} seeds, mean={mean_val:.4f}, delta={delta:+.4f} vs full-data base")
    print(f"Best: {max(scores):.4f}, Worst: {min(scores):.4f}")
    if delta >= -0.005:
        print(">>> PASS: Quality acceptable, check diversity next <<<")
    else:
        print(">>> FAIL: Quality drop too large, KILL <<<")
```

```python
# Cell 4: Strip checkpoints + zip + save to Drive
import os, torch, glob, shutil
os.chdir("/content/competition_package")
os.makedirs("logs/slim", exist_ok=True)

count = 0
for pt in sorted(glob.glob("logs/gru_parity_v1_bag85_seed*.pt")):
    basename = os.path.basename(pt)
    if '_epoch' in basename:
        continue
    ckpt = torch.load(pt, map_location="cpu", weights_only=False)
    slim = {
        "model_state_dict": ckpt["model_state_dict"],
        "config": ckpt.get("config", {}),
        "best_score": ckpt.get("best_score", None),
        "best_epoch": ckpt.get("best_epoch", None),
    }
    out = f"logs/slim/{basename}"
    torch.save(slim, out)
    orig = os.path.getsize(pt) / 1e6
    new = os.path.getsize(out) / 1e6
    print(f"{basename}: {orig:.1f}MB -> {new:.1f}MB")
    count += 1

print(f"\nStripped {count} checkpoints")

shutil.make_archive("/content/bag85_seeds", "zip",
                     "/content/competition_package/logs/slim")
sz = os.path.getsize("/content/bag85_seeds.zip") / 1e6
print(f"\nbag85_seeds.zip: {sz:.1f}MB")

shutil.copy("/content/bag85_seeds.zip",
            "/content/drive/MyDrive/wunderfund/bag85_seeds.zip")
print("Saved to Drive: MyDrive/wunderfund/bag85_seeds.zip")
```
