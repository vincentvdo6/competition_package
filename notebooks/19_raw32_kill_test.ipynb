# Raw 32 + No Normalization Kill Test (3-AI Consensus)
**Goal**: Test raw 32 features with NO z-score normalization on tightwd_v2 architecture.

**Why**: Official baseline (0.2761) uses raw features with NO normalization. Our models use z-score + 42 derived features. Z-score may be destroying regime information.

**Changes vs tightwd_v2 baseline**:
- `derived_features: false` (32 raw only, no derived)
- `normalize: false` (no z-score)
- `use_amp: false` (fp16 overflow risk with raw volumes)
- `gradient_clip: 5.0` (increased from 1.0 for raw magnitudes)

**Kill test**: 3 seeds (s42/s43/s44), pass if mean val >= 0.2550.
**Baseline**: tightwd_v2 ~0.266 val, 0.258 LB (gap -0.008)

```python
# Cell 2: Train 3 seeds with raw32 no-norm config
import os, subprocess, sys
os.chdir("/content/competition_package")

CONFIG = "configs/gru_raw32_nonorm_v1.yaml"
SEEDS = [42, 43, 44]

print(f"Raw32 No-Norm Kill Test")
print(f"Config: {CONFIG}")
print(f"Seeds: {SEEDS}")
print("=" * 60, flush=True)

for seed in SEEDS:
    print(f"\n{'='*60}")
    print(f"Training gru_raw32_nonorm_v1 seed {seed}")
    print(f"{'='*60}", flush=True)
    proc = subprocess.Popen(
        [sys.executable, "-u", "scripts/train.py",
         "--config", CONFIG,
         "--seed", str(seed), "--device", "cuda"],
        stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True
    )
    for line in proc.stdout:
        print(line, end="", flush=True)
    proc.wait()
    if proc.returncode != 0:
        print(f"ERROR: seed {seed} failed with return code {proc.returncode}")

print(f"\nAll seeds done!")
```

```python
# Cell 3: Kill test evaluation
import os, glob, torch
os.chdir("/content/competition_package")

BASELINE_VAL = 0.2580  # single tightwd_v2 LB score (LB-calibrated reference)
PASS_THRESHOLD = 0.2550  # lower threshold — prioritize generalization over val

results = []
for pt in sorted(glob.glob("logs/gru_raw32_nonorm_v1_seed*.pt")):
    basename = os.path.basename(pt)
    if '_epoch' in basename:
        continue
    ckpt = torch.load(pt, map_location="cpu", weights_only=False)
    score = float(ckpt.get("best_score", 0))
    epoch = ckpt.get("best_epoch", "N/A")
    results.append((basename, score, epoch))

print(f"{'Model':<50} {'Val Score':>10} {'Epoch':>6} {'Delta':>8}")
print("-" * 80)
for name, score, epoch in results:
    delta = score - BASELINE_VAL
    tag = "  +" if delta > 0 else "  "
    print(f"{name:<50} {score:>10.4f} {str(epoch):>6} {tag}{delta:+.4f}")

scores = [s for _, s, _ in results]
mean_val = sum(scores) / len(scores) if scores else 0
positive = sum(1 for s in scores if s > BASELINE_VAL)

print(f"\n{'='*60}")
print(f"Mean val: {mean_val:.4f} (threshold: {PASS_THRESHOLD})")
print(f"Positive seeds: {positive}/{len(scores)} (vs baseline {BASELINE_VAL})")
print(f"Delta vs baseline: {mean_val - BASELINE_VAL:+.4f}")

PASS = mean_val >= PASS_THRESHOLD
print(f"\nKILL TEST: {'PASS — proceed to LB test' if PASS else 'FAIL'}")
if not PASS:
    print("Fallback: try log1p on volumes, or lower LR, or RevIN")
```

```python
# Cell 4: Strip checkpoints + zip + save to Drive
import os, torch, glob, shutil
os.chdir("/content/competition_package")
os.makedirs("logs/slim", exist_ok=True)

for pt in sorted(glob.glob("logs/gru_raw32_nonorm*.pt")):
    basename = os.path.basename(pt)
    if '_epoch' in basename:
        continue
    ckpt = torch.load(pt, map_location="cpu", weights_only=False)
    slim = {
        "model_state_dict": ckpt["model_state_dict"],
        "config": ckpt.get("config", {}),
        "best_score": ckpt.get("best_score", None),
        "best_epoch": ckpt.get("best_epoch", None),
    }
    out = f"logs/slim/{basename}"
    torch.save(slim, out)
    orig = os.path.getsize(pt) / 1e6
    new = os.path.getsize(out) / 1e6
    print(f"{basename}: {orig:.1f}MB -> {new:.1f}MB")

# Note: no normalizer files for this config (normalize: false)
print(f"\n--- logs/slim/ contents ---")
for f in sorted(os.listdir("logs/slim")):
    sz = os.path.getsize(f"logs/slim/{f}") / 1e6
    print(f"  {f}: {sz:.1f}MB")

shutil.make_archive("/content/raw32_nonorm_kill_test", "zip",
                     "/content/competition_package/logs/slim")
sz = os.path.getsize("/content/raw32_nonorm_kill_test.zip") / 1e6
print(f"\nraw32_nonorm_kill_test.zip: {sz:.1f}MB")

shutil.copy("/content/raw32_nonorm_kill_test.zip",
            "/content/drive/MyDrive/wunderfund/raw32_nonorm_kill_test.zip")
print("Saved to Drive: MyDrive/wunderfund/raw32_nonorm_kill_test.zip")
```
